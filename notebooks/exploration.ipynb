# MNIST Exploration

This notebook explores the MNIST dataset used in the Handwritten Digit Recognition project.

## Objectives
- Load and inspect dataset shapes
- Visualize sample digits
- Establish baseline understanding before model training

```python
from importlib.util import module_from_spec, spec_from_file_location
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns

cwd = Path.cwd().resolve()
project_root = cwd if (cwd / "src").exists() else cwd.parent
loader_path = project_root / "src" / "data_loader.py"

spec = spec_from_file_location("data_loader", loader_path)
if spec is None or spec.loader is None:
    raise RuntimeError("Could not load data_loader module.")

data_loader = module_from_spec(spec)
spec.loader.exec_module(data_loader)

load_mnist_data = data_loader.load_mnist_data
```

```python
(x_train, y_train), (x_test, y_test) = load_mnist_data()

print('Train shape:', x_train.shape, y_train.shape)
print('Test shape:', x_test.shape, y_test.shape)
```

```python
sample_index = 0
sample_image = x_train[sample_index].reshape(28, 28)
sample_label = y_train[sample_index]

plt.figure()
sns.heatmap(sample_image)
plt.title(f"Sample Label: {sample_label}")
plt.axis("off")
plt.show()
```
